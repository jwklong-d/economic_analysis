```python
# import dependencies

import mplfinance as mpf
from config import api_key
import requests
import pandas as pd
import time
import datetime
from datetime import time
import matplotlib.pyplot as plt
import cufflinks as cf
import os
import time
```

```python
#input info
periodtype = input("Period type (day, month, year) or NA")
frequency = input("Frequency of data")
frequencytype = input("Frequency type (minute, daily)")
period = input("Period of data or NA")
needExtendedHoursData = input('Extended hours data (True or False)')
```

```python
#obtain list of sectors
directory = "sectors"
list_sector = []
for filename in os.listdir(directory):
    list_sector.append(filename)
```

```python
pd.read_csv('sectors/Communication Services Sector')['0'].tolist()
```

```python
for sector in list_sector:

    symbols = pd.read_csv(f'sectors/{sector}')['0'].tolist()
    

    if periodtype == "NA":
        print("Use 1")
        for symbol in symbols:


            end = "1659646252000"

            start = "962743501000"

            #url for td
            url = "https://api.tdameritrade.com/v1/marketdata/{}/pricehistory".format(symbol)

            #parameters
            paylod = {'apikey':api_key,
                      'frequencyType': frequencytype,
                      'frequency': frequency,
                      'endDate': end,
                      'startDate': start,
                      'needExtendedHoursData': needExtendedHoursData
                     }
            #requesting data
            content = requests.get(url = url, params=paylod)

            data = content.json()
            data

            open_tick = []
            high_tick = []
            low_tick = []
            close_tick = []
            volume_tick = []
            datetime_tick = []

            for i in range(len(data["candles"])):
                open_tick.append(data["candles"][i]["open"])
                high_tick.append(data["candles"][i]["high"])
                low_tick.append(data["candles"][i]["low"])
                close_tick.append(data["candles"][i]["close"])
                volume_tick.append(data["candles"][i]["volume"])
                datetime_tick.append(data["candles"][i]["datetime"])

            #converting epoch to datetime       
            file = []
            for i in range(len(datetime_tick)):

                test = datetime_tick[i]/1000 + 10000
                file.append(datetime.datetime.fromtimestamp(test))

            #convert to dataframe
            theta_df = pd.DataFrame({
                        "Open":open_tick,
                        "Close":close_tick,
                        "High":high_tick,
                        "Low":low_tick,
                        "Volume":volume_tick,
                        "Datetime": file
                })
            theta_df = theta_df.set_index("Datetime")

            theta_df["Close"] = theta_df["Close"].astype(float)

            theta_df["Open"] = theta_df["Open"].astype(float)

            theta_df["Gain%"] = (theta_df.Close - theta_df.Open)/theta_df.Open * 100

            mypath = f"{symbol}"
            if not os.path.isdir(mypath):
               os.makedirs(mypath)

            upload_file = f"{symbol}/{symbol}_{frequencytype}_{frequency}_{needExtendedHoursData}.csv"
            theta_df.to_csv(upload_file)

    else:
        print("Use 2")
        for symbol in symbols:
            #url for td
            url = "https://api.tdameritrade.com/v1/marketdata/{}/pricehistory".format(symbol)

            #parameters
            paylod = {'apikey':api_key,
                      'frequencyType': frequencytype,
                      'frequency': frequency,
                      'period' : period,
                      'periodType' : periodtype,
                      'needExtendedHoursData': needExtendedHoursData
                     }
            #requesting data
            content = requests.get(url = url, params=paylod)

            data = content.json()
            data

            open_tick = []
            high_tick = []
            low_tick = []
            close_tick = []
            volume_tick = []
            datetime_tick = []

            for i in range(len(data["candles"])):
                open_tick.append(data["candles"][i]["open"])
                high_tick.append(data["candles"][i]["high"])
                low_tick.append(data["candles"][i]["low"])
                close_tick.append(data["candles"][i]["close"])
                volume_tick.append(data["candles"][i]["volume"])
                datetime_tick.append(data["candles"][i]["datetime"])

            #converting epoch to datetime       
            file = []
            for i in range(len(datetime_tick)):

                test = datetime_tick[i]/1000 + 10000
                file.append(datetime.datetime.fromtimestamp(test))

            #convert to dataframe
            theta_df = pd.DataFrame({
                        "Open":open_tick,
                        "Close":close_tick,
                        "High":high_tick,
                        "Low":low_tick,
                        "Volume":volume_tick,
                        "Datetime": file
                })
            theta_df = theta_df.set_index("Datetime")

            theta_df["Close"] = theta_df["Close"].astype(float)

            theta_df["Open"] = theta_df["Open"].astype(float)

            theta_df["Gain%"] = (theta_df.Close - theta_df.Open)/theta_df.Open * 100

            mypath = f"{sector}"
            if not os.path.isdir(mypath):
                os.makedirs(mypath)

            upload_file = f"{sector}/{symbol}_{frequencytype}_{frequency}_{needExtendedHoursData}.csv"
            theta_df.to_csv(upload_file)

            time.sleep(0.6)
```
