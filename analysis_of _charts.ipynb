```python
#obtain list of sectors
directory = "sectors"
list_sector = []
for filename in os.listdir(directory):
    list_sector.append(filename)
```

```python
for sector in list_sector:
    
    #obtain list of sectors
    directory = sector
    list_stocks = []
    for filename in os.listdir(directory):
        list_stocks.append(filename)


    dates = []

    for name in list_stocks:

        data = pd.read_csv(f"{sector}/{name}")

        if len(data) == 0:
            continue
        data["1_day"] = data["Close"].shift(-1) 
        data["3_day"] = data["Close"].shift(-3) 
        data["5_day"] = data["Close"].shift(-5)
        data["10_day"] = data["Close"].shift(-10)
        data["20_day"] = data["Close"].shift(-20)
        data["30_day"] = data["Close"].shift(-30)
        
        data["1"] = (data["1_day"] - data["Close"]) / data["Close"] * 100
        data["5"] = (data["5_day"] - data["Close"]) / data["Close"] * 100
        data["10"] = (data["10_day"] - data["Close"]) / data["Close"] * 100
        data["20"] = (data["20_day"] - data["Close"]) / data["Close"] * 100
        data["30"] = (data["30_day"] - data["Close"]) / data["Close"] * 100


        data = data.dropna()
        
        mypath = f"Analyzing - Future/{sector}"
        if not os.path.isdir(mypath):
            os.makedirs(mypath)
        
        data.to_csv(f"{mypath}/{name}")
```
